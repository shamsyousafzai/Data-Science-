# Data visualization about agriculture of pakistan

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from bokeh.plotting import figure, output_file, show,output_notebook
output_notebook()
```

```python
from math import pi
from bokeh.palettes import Category20c
from bokeh.transform import cumsum
```

```python
def make_dashboard(x,y,z, title, file_name):
    output_file(file_name)
    p = figure(title=title, x_axis_label='Cultivated Area', y_axis_label='Total Area')
    p.line(x.squeeze(), y.squeeze(), color="firebrick", line_width=4, legend="Total Areas")
    p.line(x.squeeze(), z.squeeze(), line_width=4, legend="Cultivated Area")
    show(p)
```

```python
df = pd.read_csv("Agriculture.csv")
```

```python
df1 = df[["Cultivated Area","Uncultivated Area","Total Area"]]
df1
```

```python
x = df["Total Area"]
y = df["Cultivated Area"]
z = df["Uncultivated Area"]
```

```python
title = "Creating dashboard"
file_name = "index.html"
```

```python
make_dashboard(x,y,z,title,file_name)
```
