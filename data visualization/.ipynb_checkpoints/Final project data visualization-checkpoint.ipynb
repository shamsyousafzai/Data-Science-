# Final project data visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Topic_Survey_ASsignment.csv")
df
```

```python
df.set_index("Unnamed: 0", inplace = True)
```

```python
df.sort_values(by=['Very interested'], ascending=False)
```

```python
# total = df.loc[:,df.columns!="Unnamed: 0"]
# df = df.loc[:,df.columns!="Unnamed:0"]
df = df.astype(float)
df = (100. * df/2233).round(2)
df
```

```python
colors_list = ['#5cb85c','#5bc0de','#d9534f']
plt.title("percentage of repondents interest in data science area", fontsize =16)

result = (df.div(df.sum(1), axis=0)).plot(kind='bar',figsize=(20,8),width = 0.8,color = colors_list,edgecolor=None)
plt.xticks(fontsize=14)
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.yticks([])

for p in result.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    result.annotate('{:.0%}'.format(height), (x, y + height + 0.01))
```
