# pie chart

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("countries.csv")
```

```python
df
```

```python
df
```

```python
df.set_index("year" ,inplace = True)
```

```python
df
```

```python
df.head(10)["population"].plot(kind = "pie")
plt.title("population of first 10 years")
```

# bar_chart

```python
df.head(10)["population"].plot(kind = "bar",figsize=(10,20))
plt.title("population of first ten years",size = 20)
plt.xlabel("year",size=15)
plt.ylabel("population",size = 15)
plt.legend("year","population")
```

# Regression Plot

```python
import seaborn as sn
```

```python
sn.regplot(x = "population", y ="year" ,data=df)
```
